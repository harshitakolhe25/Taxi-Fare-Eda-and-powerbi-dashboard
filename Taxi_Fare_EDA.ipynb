```python
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include='number').columns
```

```python
for col in num_cols:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
corr = df[num_cols].corr()
corr
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.month_name()
df["Day"] = df["Date"].dt.day_name()
df["Year"] = df["Date"].dt.year
```

```python
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S")

df["Hour"] = df["Time"].dt.hour
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Booking Status", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Vehicle Type", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x="Hour", data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Booking Value"], kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Ride Distance"], kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=df["Ride Distance"])
plt.show()
```

```python
vehicle_revenue = df.groupby("Vehicle Type")["Booking Value"].mean()

vehicle_revenue.sort_values().plot(
    kind="bar",
    figsize=(8,5)
)

plt.show()
```

```python
df["Pickup Location"].value_counts().head(10).plot(
    kind="bar",
    figsize=(10,5)
)

plt.show()
```

```python
df["Drop Location"].value_counts().head(10).plot(
    kind="bar",
    figsize=(10,5)
)

plt.show()
```

```python
num_cols = df.select_dtypes(include="number")

plt.figure(figsize=(10,6))
sns.heatmap(
    num_cols.corr(),
    annot=True
)

plt.show()
```

# Key Insights

1. Auto is the most frequently booked vehicle type.
2. Peak bookings occur during evening hours.
3. Booking Value increases with Ride Distance.
4. Most cancellations are initiated by customers.
5. Certain pickup locations generate significantly more rides.
6. Booking Value contains high-value outliers.
7. Ride Distance and Booking Value show positive correlation.
